```python
df = pd.read_csv ('https://s3.sa-east-1.amazonaws.com/ckan.saude.gov.br/SIM/DO23OPEN.csv', encoding='latin1', sep=';')
```

```python
'''
cols_num são as colunas numericas do dataset, sendo o passado por dentro o indice/nome das colunas das desejadas
já as cols_cat são as colunas categoricas, e faz o mesmo precesso de busca
'''
cols_num = ['IDADE', 'PESO_NASC','QTDE_EMB']
cols_cat = ['RACACOR', 'ESTCIVIL', 'ESC', 'OCUP', 'CAUSABAS', 'LOCOCOR']
```

```python
'''
o im_num passa a função de simple imputer chamando a mediana
no im_cat passa a função do mais frequente, ou moda
essas escolhas foram feitas por conta do primeiro ser um numerico e poder realizar a mediana
enquanto quando é texto não se torna possivel realizar isso, então passa o que mais apareceu
'''
im_num = SimpleImputer(strategy='median')
im_cat = SimpleImputer(strategy='most_frequent')
```

```python
for col in cols_num:
    if col in df.columns:
        df[col] = im_num.fit_transform(df[[col]])
```

```python
cols_cat_to_encode = ['RACACOR', 'ESC', 'OCUP', 'CAUSABAS']
enc = OneHotEncoder(handle_unknown='ignore', sparse_output='False')
```

```python
df_processed = pd.concat([df.drop(columns=cols_cat_to_encode).reset_index(drop='True')], axis=1)
```

```python
cols_num_to_escale = ['IDADE']
scaler = StandardScaler()
```

```python
df_processed[cols_num_to_escale] = scaler.fit_transform(df_processed[cols_num_to_escale])
```

```python
x = df_processed.drop(columns=['ORIGEM', 'TIPOBITO', 'SEXO'])
y = df_processed['SEXO']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```
